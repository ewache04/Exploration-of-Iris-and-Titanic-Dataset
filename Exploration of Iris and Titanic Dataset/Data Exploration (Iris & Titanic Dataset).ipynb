# Task 1

Task 1: Comparing Visualization Techniques for Multivariate Data (Iris Dataset)")
Introductory Remarks
In this task, we will investigate the Iris dataset and compare different visualization techniques
in order to comprehend the relationships between iris species based on sepal and petal measurements.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure format for better quality on high-DPI displays
%config InlineBackend.figure_format='retina'

# Set seaborn style to darkgrid (you can change this style if needed)
sns.set(style="darkgrid")

# Set default figure size and label padding
plt.rcParams['figure.figsize'] = (9, 6)
plt.rcParams['axes.labelpad'] = 10
```

```python
#Load dataset
df_iris = sns.load_dataset("iris")
```

```python
print("Iris Dataset Summary:")
# Inspect the columns of the DataFrame
print("Columns in the DataFrame:")
print(df_iris.columns)

```

```python
# Display the first few rows of the DataFrame
print("\nFirst few rows of the DataFrame:")
print(df_iris.head())

```

```python
# Display the last few rows of the DataFrame
print("\nLast few rows of the DataFrame:")
print(df_iris.tail())
```

```python
print("\nDescription of the DataFrame:")
print(df_iris.describe())
```

```python
# Visualization Techniques
print("\nVisualization Techniques:")
print("1. Scatterplot Matrix")
print("2. Bar Chart")
print("3. Parallel Coordinates Plot")
```

```python
# Visualization 1: Scatterplot Matrix
# The first visualization in this study is a scatterplot matrix.

sns.pairplot(df_iris, hue='species')
plt.title("Scatterplot Matrix of Iris Dataset")
plt.show()

# Advantages:
print("Advantages of Scatterplot Matrix:")
print("- Facilitates the identification of patterns and correlations among pairs of variables.")
print("- Enables exploration of the correlation between sepal and petal measures and various iris species.")
print("- Provides a comprehensive perspective on multivariate data, allowing concurrent analysis of many factors.")
print("- Helps examine data point distribution and facilitates the identification of potential outliers.")

# Disadvantages:
print("\nDisadvantages of Scatterplot Matrix:")
print("- Can become complex and challenging to comprehend with a substantial number of variables.")
print("- More suitable for visualizing relationships between continuous variables; less effective for categorical data.")
print("- Visualizing categorical data using scatterplots can be challenging and less intuitive compared to other data types.")
```

```python
# Visualization 2: Bar Chart
# The second visualization in this study is a bar chart.

plt.figure(figsize=(8, 6))
sns.countplot(x='species', data=df_iris)
plt.title("Count of Iris Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()

# Advantages:
print("Advantages of Bar Chart:")
print("- Simplicity: Bar charts possess a plain and comprehensible nature, making them suitable for communicating uncomplicated data, such as numerical counts.")
print("- Clear Comparison: The tool enables clear comparisons among several categories, specifically in this instance, the counts of distinct iris species.")
print("- Highly effective for visually representing categorical data.")

# Disadvantages:
print("\nDisadvantages of Bar Chart:")
print("- Limited Effectiveness for Multivariate Data: Bar charts focus on a single variable, making them less effective for exploring multivariate data.")
print("- Limited Representation of Continuous Variables: The technique has limitations in representing relationships and patterns among continuous variables when compared to alternative methods.")
```

```python
# Visualization 3: Parallel Coordinates Plot
# The third visualization in this study is a parallel coordinates plot.

from pandas.plotting import parallel_coordinates
plt.figure(figsize=(10, 8))
parallel_coordinates(df_iris, 'species', colormap='coolwarm')
plt.title("Parallel Coordinates Plot of Iris Dataset")
plt.show()

# Advantages:
print("Advantages of Parallel Coordinates Plot:")
print("- Powerful tool for displaying and discerning patterns in multivariate data within high-dimensional space.")
print("- Enables examination of relationships between variables and observation of interconnectedness and data point movement along axes.")
print("- Appropriate for both categorical and continuous data, effectively managing and analyzing variables of both types.")

# Disadvantages:
print("\nDisadvantages of Parallel Coordinates Plot:")
print("- Complexity of interpretation can pose a significant challenge, especially with numerous variables and complex datasets.")
print("- Limited effectiveness for tiny datasets; may not provide as comprehensive insights as simpler techniques.")
print("- Overlapping lines occur when variables share common characteristics, making differentiation challenging and potentially causing confusion.")
```

```python
# Conclusion for Task 1
print("\nTask 1 Conclusion:")
print("In this exploration of the Iris dataset using different visualization techniques, we have gained valuable insights into the relationships between iris species based on sepal and petal measurements.")

# Scatterplot Matrix
print("\nScatterplot Matrix:")
print("The scatterplot matrix has proven to be a powerful tool for understanding the dataset. It allows us to identify intricate patterns and correlations among pairs of variables. By examining how sepal and petal measurements relate to various iris species, we can discern distinct clusters and patterns, aiding in species differentiation. This technique excels at providing multivariate insights and revealing the distribution of data points. Additionally, it assists in the identification of potential outliers, which can be crucial for data quality assessment.")

# Bar Chart
print("\nBar Chart:")
print("The bar chart, on the other hand, offers a different perspective. While it may not provide the depth of multivariate insights that the scatterplot matrix offers, it excels in simplifying the representation of data. Through this chart, we can easily compare the counts of distinct iris species. Its simplicity makes it a valuable tool for straightforward data communication. Bar charts are particularly effective for visualizing categorical data, as they provide a clear visual representation of categorical counts, aiding in quick comparisons and assessments.")

# Parallel Coordinates Plot
print("\nParallel Coordinates Plot:")
print("Lastly, the parallel coordinates plot offers yet another layer of understanding. It is a powerful tool for displaying and discerning patterns in multivariate data within high-dimensional space. By examining the relationships between variables, we can observe how these variables are interconnected and how data points exhibit movement along axes. What sets this technique apart is its versatility; it is suitable for both categorical and continuous data. However, the complexity of interpretation can pose a challenge, especially with a multitude of variables and intricate datasets. Additionally, in the case of tiny datasets, it may not yield as comprehensive insights as simpler techniques. Care must also be taken to address the phenomenon of overlapping lines, which can make differentiation challenging and potentially lead to confusion.")

# Overall
print("\nOverall, the choice of visualization technique should align with the specific objectives of the analysis. Each technique has its strengths and weaknesses, and the selection should be based on the characteristics of the dataset and the insights sought. In this study, we have harnessed the power of scatterplot matrices for in-depth exploration, bar charts for simplicity and clear counts, and parallel coordinates plots for high-dimensional insights.")
```

```python
# Data Loading
df_titanic = sns.load_dataset("titanic")
```

```python
# Exploration of dataset
print("Titanic Dataset Summary:")
```

```python
# Inspect the columns of the DataFrame
print("Columns in the DataFrame:")
print(df_titanic.columns)
```

```python
# Display the first few rows of the DataFrame
print("\nFirst few rows of the DataFrame:")
print(df_titanic.head())
```

```python
# Display the last few rows of the DataFrame
print("\nLast few rows of the DataFrame:")
print(df_titanic.tail())
```

```python
# Discribe the DataFrame
print("\nDescribe the DataFrame:")
print(df_titanic.describe())
```

```python
# Understanding the Narrative
print("\nUnderstanding the Narrative:")
print("Factors affecting survival rates on the Titanic")
```

```python
# Visualization 1: Survival by Passenger Class
plt.figure(figsize=(8, 6))
sns.barplot(x='class', y='survived', data=df_titanic, errcolor='0.2')  # Use errcolor to specify error bar color
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()

print("Part One of the Story: Passenger Class and Survivability")
print("\nFirst Visualization: Passenger Class Survival")
print("We start our journey by investigating how passenger class affects survival.")
print("We use an eye-catching bar chart to convey this data.")
print("The graphic makes clear how drastically different passenger classes' survival percentages differ from one another.")
print("Third-class passengers had the lowest chances of surviving, while First-class and Second-class travelers had the best survival rates.")

print("\nFirst Discussion:")
print("We explore the human tales that lie behind the statistics.")
print("In addition to statistics, the class-based survival differences provide insight on the socioeconomic realities of the period.")
print("It makes us stop and consider the injustice of lifeboat distribution and the courage of those who faced such extreme odds.")
```

```python
# Visualization 2: Survival by Gender
plt.figure(figsize=(8, 6))
sns.barplot(x='sex', y='survived', data=df_titanic, errcolor='0.2')
plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.show()

print("Second Narrative Element: Women and the Sea's Chivalry")
print("\nVisualization 2: Gender-Based Survival")
print("We continue our story as we investigate how gender affects survival.")
print("Women's survival rates were much greater than men's, as shown by an impactful bar chart.")
print("Women and children were granted preferential access to lifeboats during the evacuation, and the depiction honors their bravery and gallantry.")

print("\nDiscussion 2:")
print("Tales of bravery and selflessness are evoked by the differences in survival based on gender.")
print("We reflect on the deeds of generosity and compassion that took place in the face of tragedy.")
print("It is this story element that embodies the human spirit in the face of hardship.")
```

```python
# Visualization 3: Age Distribution of Passengers
plt.figure(figsize=(10, 6))
sns.histplot(df_titanic['age'], bins=20, kde=True)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

print("Third Story Element: Getting Older and the Young Souls")
print("\nVisualization 3: Passenger Age Distribution")
print("We take a moment as our story progresses to look at the age range of Titanic passengers.")
print("A poignant histogram indicates that a substantial portion of the travelers were young adults.")
print("This background becomes important when we investigate patterns of survival related to age.")

print("\nDiscussion 3:")
print("The age distribution emphasizes the high percentage of youthful passengers and sets the scene for our story.")
print("As we investigate how age affected survival outcomes, it becomes pertinent.")
print("This feature allows us to empathize with the aspirations and hopes of the younger travelers.")
```

```python
# Visualization 4: Survival by Family Size
# Calculate survival counts by alone
survival_counts = df_titanic['alone'][df_titanic['survived'] == 1].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(survival_counts, labels=survival_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Survival Rate by Traveling Alone or with Companions")
plt.show()

print("Fourth Narrative Element: Alone or Not Alone")
print("\nVisualization 4: Survival by Traveling Alone")
print("Our narrative takes a turn as we explore how traveling alone or with companions influenced survival.")
print("An emotive bar chart demonstrates that passengers traveling alone had different survival rates,")
print("possibly due to the choices they made during the evacuation.")
print("\nDiscussion 4:")
print("The impact of traveling alone or with companions unravels stories of choices made during a time of chaos.")
print("It prompts us to imagine the difficult decisions faced by individuals in such situations.")
```
